 step 1-pandas-load and analyze data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Step 2-read the csv file into a data frame and loaded data sccuessfully

```python
df = pd.read_csv('/content/student-mat.csv', sep=';')
print("Dataset Loaded Successfully")
```

```python
import zipfile
import os

zip_file_path = '/content/student.zip'
extract_to_path = '/content/'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_path)

print(f"'{zip_file_path}' unzipped to '{extract_to_path}' successfully.")
print("Contents after unzipping:")
print(os.listdir(extract_to_path))
```

Step 3-check rows and columns

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
print("\nDataset Shape:")
print(df.shape)
print("\nColumns:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
print("\nStatistical Summary:")
print(df.describe())
```

step 5- checking the missing values


 if the output is o then meaning is no missing values









```python
print("\nMissing Values:")
print(df.isnull().sum())
```

Step 6-check duplicates and remove duplicates
keep only unique data

```python
duplicates = df.duplicated()
print("\nDuplicate Rows:")
print(duplicates.sum())
df=df.drop_duplicates()
print("Duplicates Removed")
```

question 1-Average final grade G

```python
avg_grade = df["G3"].mean()
print("\nAverage Grade:")
print(avg_grade)
```

Question 2-number of students scored above 15

```python
above_15 = df[df["G3"]>15]
print("\nStudents with Grade Above 15:")
print(above_15)
```

Question 3-finding study time correlated with perfomance

```python
correlation =df["studytime"].corr(df["G3"])
print("\nCorrelation between Study Time and Grade:")
print(correlation)
```

Question 4 -finding which gender performs better on average

```python
gender_avg=df.groupby("sex")["G3"].mean()
print("\nAverage Grade by Gender:")
print(gender_avg)
```

Visualization of data
Histogram of Grades

```python
plt.figure(figsize=(8,5))
plt.hist(df["G3"])
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("number of students")
plt.show()
```

Scatterplot: study time vs grades.

```python
plt.figure(figsize=(8,5))
plt.scatter(
    x=df["studytime"],
    y=df["G3"]
)
plt.title("Study Time vs Grade")
plt.xlabel("Study Time")
plt.ylabel("Grade")
plt.show()
```

Bar chart: male vs female average score

```python
plt.figure(figsize=(6,5))
gender_avg.plot(kind="bar")
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

FINAL CONCLUSION

```python
print("\n=======conclusion========")
print("Average Grade:",avg_grade)
print("Students with Grade Above 15:")
print(above_15)
print("Correlation between Study Time and Grade:",correlation)
print("Average Grade by Gender:")
print(gender_avg)
```
